```python
URLS=[
  'https://en.wikipedia.org/wiki/Data_science',
  'https://en.wikipedia.org/wiki/Software_development',
  'https://en.wikipedia.org/wiki/Hugging_Face'
]
loader=UnstructuredURLLoader(urls=URLS)
data=loader.load()
```

```python
len(data)
```

```python
text_spilter=CharacterTextSplitter(separator='\n',chunk_size=1000,chunk_overlap=200)
text_chunks=text_spilter.split_documents(data)
len(text_chunks)
```

```python
text_chunks[1]
```

```python
embeddings=GoogleGenerativeAIEmbeddings(model='models/embedding-001')
```

```python
query_result=embeddings.embed_query("Hello World")
len(query_result)
```

```python
query_result
```

```python
vectorstore=FAISS.from_documents(text_chunks,embeddings)
```

```python
model_name = 'gemini-1.5-flash'
```

```python
llm=ChatGoogleGenerativeAI(model=model_name)
```

```python
response=llm.predict('What is the roadmap of Full stack developer?').strip()
print(response)
```

```python
chain=RetrievalQAWithSourcesChain.from_llm(llm=llm,retriever=vectorstore.as_retriever())
```

```python
result=chain({'question':'What is Hugging face and how much models in Hugging Face?'},return_only_outputs=True)
```

```python
result['answer']
```

```python
while True:
  query=input(f"Prompt:")
  if query=='exit':
    print('Existing')
    sys.exit()
  if query == '':
    continue
  result=chain({'question':query})
  print(f"Answer:"+result["answer"])
```
